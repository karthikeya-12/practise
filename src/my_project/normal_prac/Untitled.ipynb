```python
from sqlalchemy import create_engine, Column, Integer, VARCHAR, Text, Boolean, select
from sqlalchemy.orm import declarative_base, sessionmaker
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
import os
```

```python
engine = create_engine(
    f"postgresql+asyncpg://{os.getenv('DB_USER')}:{os.getenv('DB_PASSWORD')}"
    f"@{os.getenv('DB_HOST')}/{os.getenv('DB_NAME')}",
    echo=False,
)
```

```python
Base = declarative_base()
```

```python
class ChatBot(Base):
    __tablename__ = "chat_bot"
    __table_args__ = {"schema": "ai_compute_modeller"}

    chat_bot_id = Column(Integer, primary_key=True)
    history_id = Column(VARCHAR)
    user_id = Column(Integer)
    chat_title = Column(Text)
    status = Column(Boolean)
    created_by = Column(VARCHAR)
    updated_by = Column(VARCHAR)
```

```python
from contextlib import contextmanager
Session = sessionmaker(bind=engine)

@contextmanager
def get_db():
    with Session() as session:
        try:
            yield session
            session.commit()
        except Exception:
            session.rollback()
        finally:
            session.close()
```

```python
with get_db() as session:
    result = session.query(ChatBot).filter_by(user_id=220)
    if result:
        print(result.chat_bot_id)
```
